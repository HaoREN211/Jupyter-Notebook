```python
import numpy as np
import pandas as pd
import time
```

```python
N = []
for i in range(7):
    N.append([1]*7)

for i in range(7):
    N[i][i]=0
    
N = pd.DataFrame(N)
```

```python
P = N
for i in range(np.shape(N)[0]):
    sum_row = 0.0
    for j in range(np.shape(N)[1]):
        sum_row += N.iloc[i, j]
    for j in range(np.shape(N)[1]):
        P.iloc[i, j] = round(N.iloc[i, j]*100/sum_row)/100
        
matrix_r = np.identity(7)

v_final = []
a = 0.15
start = time.perf_counter()
for j in range(7):
    r0 = np.array(matrix_r[j])
    v = r0
    Pt = (P.T).values
    for i in range(100):
        v = (1-a)*np.dot(Pt, v)+a*r0
    for i in range(np.size(v)):
        v[i] = round(v[i]*100)/100
    v_final.append(v)
elapsed = (time.perf_counter() - start)
print("Time used:",elapsed)
v_final = pd.DataFrame(v_final)
v_final = v_final.T
v_final
```
